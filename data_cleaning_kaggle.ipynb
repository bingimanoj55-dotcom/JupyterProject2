```python
df['Gender'] = df['Gender'].str.lower()
df['Gender'].value_counts()
```

```python
# Do NOT use inplace=True when assigning back
df['Gender'] = df['Gender'].replace({
    'm': 'male',
    'f': 'female'
})

# Now recompute counts
print(df['Gender'].value_counts())
```

```python
df['Country'].unique()
```

```python
print("Before:\n", df['Country'].unique())
df['Country'] = df['Country'].replace({
    'USA': 'United States',
    'UK': 'United Kingdom',
    'UAE': 'United Arab Emirates',
    'Indiaa': 'India',
    'INDIA': 'India',
    'norway': 'Norway',
    'Rsa': 'Russia'
})

print("After:\n", df['Country'].unique())
```

```python
df['Residence'].unique()

df['Residence'] = df['Residence'].replace({
    'BI-Residence' : 'BI Residence',
    'BIResidence' : 'BI Residence',
    'BI_Residence' : 'BI Residence'
})

print("After:\n", df['Residence'].unique())
```

```python
print("Before:\n", df['Prev_Education'].unique())
df['Prev_Education'] = df['Prev_Education'].replace({
    'HighSchool': 'High School',
    'Barrrchelors': 'Bachelors',
    'Diplomaaa': 'Diploma',
    'diploma': 'Diploma',
    'DIPLOMA': 'Diploma'
})

print("After:\n", df['Prev_Education'].unique())
```
