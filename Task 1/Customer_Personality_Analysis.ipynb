```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("C:/Users/bingi/Downloads/marketing_campaign.csv")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.describe()
```

```python
print("Total no.of rows :" ,df.shape[0])
print("Total no.of columns :" ,df.shape[1])
```

```python
print("Total values are missing in the columns : ",df.isnull().any(axis=0).sum())
print("Total values are missing in the rows :\n",df.isnull().sum())
```

```python
df['Income'].fillna(df['Income'].mean(),inplace = True)
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df['Income'].mean().round(2)
```

```python
df['Marital_Status'].value_counts()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone' : 'Single',
    'YOLO' : 'Single',
})
```

```python
df['Marital_Status'].value_counts()
```

```python
df.drop(df[df['Marital_Status'] == 'Absurd'].index,inplace = True)
df['Marital_Status'].value_counts()
```

```python
df['Education'].value_counts()
```

```python
df['Education'] = df['Education'].replace({
    '2n Cycle' : 'Master'
})
df['Education'].value_counts()
```

```python
df['HouseHold'] = df['Kidhome'] + df['Teenhome']
df['Total_spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df['Total_purchases'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
df['Campaign_response'] = df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']].max(axis=1)
df.head()
```

```python
df['Age'] = pd.to_datetime('today').year - df['Year_Birth']
df.head()
```

```python
purchase = df.groupby('Total_spending')['Total_purchases'].sum().reset_index()
df.loc[purchase['Total_purchases'].idxmax()]
```

```python
plt.figure(figsize=(12,7))
sns.histplot(data = df , x = 'Age' , bins = 10 , color = 'Blue' , edgecolor = 'White',kde = True)
plt.title('Customer Age Distribution' , fontsize = 18 , fontweight = 'bold' , pad = 20)
plt.xlabel('Age(Years)')
plt.ylabel('Count')
plt.grid(True, alpha=0.15, linestyle='--')
plt.show()
```

```python
import plotly.express as px

fig = px.histogram(
    df, 
    x='Age', 
    nbins=10, 
    title='Customer Age Distribution', 
    color_discrete_sequence=['blue']
)

fig.show()
```

```python
import matplotlib.pyplot as plt

status_counts = df['Marital_Status'].value_counts()

# Create pie chart
plt.figure(figsize=(8,5))
plt.pie(
    status_counts.values, 
    labels=status_counts.index, 
    colors=['red','blue','green','orange','purple'], 
    autopct='%1.1f%%', 
    startangle=90
)

plt.title('Customer Marital Status Distribution', fontsize=18, fontweight='bold', pad=20)
plt.show()
```

```python
status_counts = df['Marital_Status'].value_counts().reset_index()
status_counts.columns = ['Marital_Status', 'Count']

plt.figure(figsize=(12,7))
sns.barplot(data=status_counts, x='Marital_Status', y='Count', palette='magma')

plt.title('Customer Marital Status Distribution', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.grid(True, alpha=0.15, linestyle='--')
plt.show()

```

```python
import plotly.express as px

income_for_marital_status = df.groupby('Marital_Status')['Income'].sum().reset_index()

fig = px.bar(income_for_marital_status,
             x='Marital_Status',
             y='Income',
             color='Marital_Status',
             title='Total Income by Marital Status')

fig.update_layout(
    xaxis_title='Marital Status',
    yaxis_title='Total Income',
    title_font=dict(size=18, family='Arial', color='black'),
    plot_bgcolor='rgba(0,0,0,0)',
    xaxis=dict(showgrid=True, gridcolor='lightgrey', gridwidth=0.5, griddash='dash'),
    yaxis=dict(showgrid=True, gridcolor='lightgrey', gridwidth=0.5, griddash='dash')
)

fig.show()
```

```python
income_on_education = df.groupby('Education')['Income'].sum().reset_index()

fig = income_on_education.pivot_table(
                    index='Education',
                    values='Income',
                    aggfunc='sum',
                    fill_value=0)
fig
```

```python
Avg_income_by_education = df.groupby('Education')['Income'].mean().reset_index()

fig = px.bar(Avg_income_by_education,
             x = 'Education',
             y = 'Income',
             color = 'Education',
             title = 'Average Income by Education Level')
fig.show()
```

```python
income_on_education = df.groupby('Education')['Income'].sum().reset_index()

fig = px.bar(income_on_education,
             x='Education',
             y='Income',
             color = 'Education',
                title='Total Income by Education Level')
fig.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Income', hue='Education')
plt.title('Age Vs Income')
plt.show()
```

```python
categorical_cols = ['MntWines','MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

sns.heatmap(df[categorical_cols].corr(), annot=True, cmap="YlGnBu")
plt.title('Correlation Heatmap of Spending Categories', fontsize=18, fontweight='bold', pad=20)
plt.show()
```

```python
Total_purchases = ['NumDealsPurchases' ,'NumWebPurchases' , 'NumCatalogPurchases' , 'NumStorePurchases']

sns.heatmap(df[Total_purchases].corr(), annot = True , cmap="YlGnBu")
plt.title('Correlation Heatmap of Purchase Types', fontsize=18, fontweight='bold', pad=20)
plt.show()
```

```python
numerical_cols = ['Age', 'Income', 'Total_spending', 'Total_purchases', 'Recency']

plt.figure(figsize=(8,6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap="YlGnBu")
plt.title('Correlation Heatmap of Numerical Features', fontsize=18, fontweight='bold', pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
