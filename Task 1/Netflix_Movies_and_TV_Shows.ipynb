```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("C:\\Users\\bingi\\Downloads\\netflix_titles.csv\\netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print('Number of rows after dropping duplicates:', df.shape[0])
print('Number of columns after dropping duplicates:', df.shape[1])
```

```python
df['type'].value_counts()
```

```python
df.tail()
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df[df['duration'].isnull()]
```

```python
df.drop(df[df['duration'].isnull()].index, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df['director'] = df['director'].fillna('Unknown director')
df['cast'] = df['cast'].fillna('Unknown cast')
df['country'] = df['country'].fillna('Unknown country')
df['date_added'] = df['date_added'].fillna('Unknown date')
df['rating'] = df['rating'].fillna('No rating')
df.isnull().sum()
```

```python
type_counts = df['type'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x = type_counts.index , y = type_counts.values , palette='viridis')
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
country_wide = df['country'].value_counts().head(10).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
sns.barplot(x = country_wide.values , y = country_wide.index , palette='magma')
plt.title('Top 10 Countries with the Most Netflix Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
release_year_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x = release_year_counts.index , y = release_year_counts.values , marker='o')
plt.title('Number of Netflix Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
country_type = df.groupby(['country', 'type']).size().reset_index(name='Count').head(10)

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=country_type, x='country', y='Count', hue='type', palette='viridis')
plt.title('Country-wise Distribution of Movies and TV Shows on Netflix')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Type')
plt.show()
```

```python
top_10_countries_by_movies = df[(df['type'] == 'Movie') & (df['country'] != 'Unknown country')]['country'].value_counts().head(10).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
sns.barplot(x = top_10_countries_by_movies.values , y = top_10_countries_by_movies.index , palette='coolwarm')
plt.title('Top 10 Countries with the Most Movies on Netflix')
plt.xlabel('Number of Movies')
plt.ylabel('Country')
plt.show()
```

```python
top_10_countries_by_TV_Shows = (
    df[(df['type'] == 'TV Show') & (df['country'] != 'Unknown country')]
    ['country']
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=top_10_countries_by_TV_Shows.values,
    y=top_10_countries_by_TV_Shows.index,
    palette='coolwarm'
)
plt.title('Top 10 Countries with the Most TV Shows on Netflix')
plt.xlabel('Number of TV Shows')
plt.ylabel('Country')
plt.show()

```

```python
top_rating = df['rating'].value_counts().head(10).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
sns.barplot(x = top_rating.values , y = top_rating.index , palette='viridis')
plt.title('Top 10 Ratings on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Rating')
plt.show()
```

```python
genere = df['listed_in'].str.split(',').explode().str.strip().value_counts().head(10).sort_values(ascending=True)
genere.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.show()
```
