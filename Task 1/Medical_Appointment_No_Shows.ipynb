```python
def rename_columns(df):
    df.rename(columns={
        'PatientId': 'Patient_id',
        'AppointmentID': 'Appointment_id',
        'ScheduledDay': 'Scheduled_day',
        'AppointmentDay': 'Appointment_day',
        'No-show': 'No_show',
    }, inplace=True)
    return df
df = rename_columns(df)
df.head()
```

```python
df['No_show'] = df['No_show'].map({'No' : 0, 'Yes' : 1})
df.head()
```

```python
def rename_columns(df):
    return df.rename(columns = {
        'Hipertension': 'Hypertension',
        'Handcap': 'Handicap',
        'No_show': 'Showed'
    })
df = rename_columns(df)
df.head()
```

```python
Gender_status = df['Gender'].value_counts().reset_index()
Gender_status.columns = ['Gender', 'Count']  

plt.figure(figsize=(8,6))
sns.barplot(data=Gender_status, x='Gender', y='Count', color='blue')
plt.title('Gender Distribution', fontsize=18, fontweight='bold')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
Scholarship_status = df['Scholarship'].value_counts().reset_index()
Scholarship_status.columns = ['Scholarship', 'Count']

fig = px.bar(Scholarship_status,
             x='Scholarship',
             y='Count',
             color='Scholarship',
             title='Scholarship Distribution')

fig.update_layout(
    xaxis_title="Scholarship",
    yaxis_title="Count",
    title_font=dict(size=18, family='Arial', color='black')
)

fig.show()
```

```python
disease_status = df.groupby('Gender')[['Hypertension', 'Diabetes', 'Alcoholism', 'Handicap' , 'Showed']].sum().reset_index()

fig = pd.pivot_table(disease_status,
                     index = 'Gender',
                     values = ['Hypertension', 'Diabetes', 'Alcoholism', 'Handicap', 'Showed'],
                     aggfunc = 'sum',
                     fill_value = 0)

fig['Grand_Total'] = fig.sum(axis=1)
fig.loc['Total'] = fig.sum(axis=0)
fig
```

```python
disease_status = df.groupby('Gender')[['Hypertension', 'Diabetes', 'Alcoholism', 'Handicap' , 'Showed']].mean().reset_index()

disease_status.set_index('Gender').plot(kind='bar', figsize=(10,6))

plt.title('Average Disease Indicators by Gender')
plt.ylabel('Mean Value')
plt.xlabel('Gender')
plt.legend(title='Disease')
plt.show()
```

```python
disease_status = df.groupby('Gender')[['Hypertension', 'Diabetes', 'Alcoholism', 'Handicap' , 'Showed']].sum().reset_index()

disease_status.set_index('Gender').plot(kind='bar', figsize=(10,6))

plt.title('Count of Disease Indicators by Gender')
plt.ylabel('Sum Value')
plt.xlabel('Gender')
plt.legend(title='Disease')
plt.show()
```

```python
show = df.Showed == 1
Nshow = df.Showed == 0
```

```python
def attend (df, age , show, Nshow):
    plt.figure(figsize=[16,5])
    df[age][show].hist(alpha = .5 , bins = 10, color ='r')
    df[age][Nshow].hist(alpha = .5 , bins = 10, color ='g')
    plt.legend();
    plt.title("comperision age & patient_show")
    plt.xlabel("Age")
    plt.ylabel("PatientNum")
    plt.show()
    
attend(df, "Age" ,show , Nshow) 
```

```python
plt.figure(figsize=[12, 5])

# Plotting mean age for patients who showed up
df[show].groupby(['Hypertension', 'Diabetes'])['Age'].mean().plot(kind='bar', color='Red', alpha=0.7, label='Show')

# Plotting mean age for patients who did not show up
df[Nshow].groupby(['Hypertension', 'Diabetes'])['Age'].mean().plot(kind='bar', color='Blue', alpha=0.7, label='No Show')

plt.legend()
plt.title("Comparison of Hypertension and Diabetes to Mean Age of Patients")
plt.xlabel("Hypertension, Diabetes Status")
plt.ylabel("Mean Age")
plt.xticks(rotation=0)
plt.show()
```
